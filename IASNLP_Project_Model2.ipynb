```python
# Use sklearn support vector regression to predicit our data:
from sklearn.svm import SVR
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
train_f = pd.read_csv("Testing_500_final.csv")
ftr_score = train_f.iloc[:,9:12].to_numpy()
scores = train_f.iloc[:,7:8].to_numpy().ravel()

X_train,X_test,y_train,y_test = train_test_split(ftr_score,scores,test_size=0.33,random_state=42)
#Convert to 1d Vector


svr_rbf = SVR(kernel= 'rbf', C= 1e3, gamma= 0.1)
svr_rbf.fit(X_train, y_train)
pre = svr_rbf.predict(X_test)

```

```python
def accuracy(y_test,pred,pct):
  n = len(y_test)
  tol_Delta = abs(pct*y_test)
  abs_Delta = abs(pred - y_test)

  results = abs_Delta < tol_Delta
  acc = sum(results)/n
  return acc
```
